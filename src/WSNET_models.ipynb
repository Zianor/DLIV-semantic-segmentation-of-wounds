```python
import os

os.environ["SM_FRAMEWORK"] = "tf.keras"

from src.WSNET.wsnet import train_model
import matplotlib.pyplot as plt

import numpy as np
test_images_count = 403
BATCH_SIZE = 16
```

## Model Linknet Local
(https://github.com/subbareddy248/WSNET/blob/main/Code/Medical_CNN_Linknet-Local.py)

```python
model_linknet_local, train_gen_linknet_local, val_gen_linknet_local, test_gen_linknet_local = train_model(segmentation_model="linknet", model_architecture="local", load_only=True)
```

```python
results = model_linknet_local.evaluate(test_gen_linknet_local, steps=np.ceil(float(test_images_count) / float(BATCH_SIZE)))
dict(zip(model_linknet_local.metrics_names, results))
```

```python
for image, label in test_gen_linknet_local:
    res = model_linknet_local.predict(image)
    plt.subplots(BATCH_SIZE, 3, figsize=(15,15))
    for i in range(BATCH_SIZE):
        plt.subplot(BATCH_SIZE,3,i*3 + 1)
        plt.imshow(image[i, :, :, :])
        plt.axis('off')
        plt.subplot(BATCH_SIZE,3, i*3 + 2)
        plt.imshow(res[i, :, : , :], cmap='gray')
        plt.axis('off')
        plt.subplot(BATCH_SIZE,3, i*3 + 3)
        plt.imshow(label[i, :, : , :], cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.show()
    break
```
